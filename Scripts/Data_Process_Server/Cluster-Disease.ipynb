## Autoencoder, reduce 31941 features to 128 features

```python
from keras.models import Sequential, Model
from keras import layers
from keras import optimizers
from keras.layers import  BatchNormalization, Activation,  Dense, Input, Dropout
from sklearn.preprocessing import minmax_scale
```

```python
prefix = "/data/liu/mimic3/USER_VECTORS"
df_final = read_data(join(prefix, 'four_combined_features'))
df_final = df_final.set_index("SUBJECT_ID")
df_final.head()
```

```python
## Scale data to [0,1]
# df_final = df_final.set_index("SUBJECT_ID")
df_final_scaled = minmax_scale(df_final, axis = 0)
df_final_scaled[:10]
```

```python
## Build autoencoder model
model = Sequential()
model.add(Dense(5000,  activation='relu', input_shape=(31941,)))
model.add(BatchNormalization())
model.add(Dense(2500, activation='relu'))
model.add(Dense(128,  activation='linear', name="bottleneck"))
model.add(BatchNormalization())
model.add(Dense(2500, activation='relu'))
model.add(Dense(5000,  activation='relu'))
model.add(Dense(31941,  activation='softmax'))
model.summary()
```

```python
df_final_scaled.shape
```

```python
encoder = Model(model.input, model.get_layer('bottleneck').output)
model.compile(loss='mse', optimizer = optimizers.Adam())
model.fit(df_final_scaled,df_final_scaled, batch_size=5000, epochs=3, shuffle=True)
```

### Kmeans Cluster on 128 extracted features from Autoencoder

```python
from sklearn.cluster import KMeans
from collections import Counter
import numpy as np
from MulticoreTSNE import MulticoreTSNE as TSNE
from os.path import join
import matplotlib.pyplot as plt
import random

from tools import *
```

```python
write_pre = "/data/liu/mimic3/EXTRACT_FEATURES"
repre_128 = read_data(join(write_pre, "Autoencoder_128"))
```

```python
random.seed(2019)
kmeans = KMeans(n_clusters=4)
kmeans.fit(repre_128.iloc[:,1:])
Counter(kmeans.labels_)
```

```python
X_embedded = TSNE(n_jobs=4,n_components=2).fit_transform(repre_128.iloc[:,1:])
X_embedded.shape
```

```python
plt.figure('K-means with 3 clusters')
plt.scatter(X_embedded[:, 0], X_embedded[:, 1], c=kmeans.labels_)
plt.show()
```

```python
Autoencoder_128_2d = pd.DataFrame(X_embedded)
Autoencoder_128_2d = pd.concat([repre_128[['SUBJECT_ID']], Autoencoder_128_2d], axis=1)
write2file(Autoencoder_128_2d, join(write_pre, "Autoencoder_128_2d"))
Autoencoder_128_2d.head()
```

```python
## Save kmeans cluster labels into the file
cluster_pre = "/data/liu/mimic3/CLUSTER"
write2file(pd.concat([repre_128[['SUBJECT_ID']], pd.DataFrame(kmeans.labels_)], axis=1), \
               join(cluster_pre, "kmeans_C4_F128"))
```

```python
def plot_bar_x(label, data, label_name=['NDC','Number of Patients'], title = "Number of Patients Taking Each Prescription"):
    fig, ax = plt.subplots(figsize=(20,10))
    # this is for plot ting purpose
    x_label, y_label = label_name
    index = np.arange(len(label))
    plt.bar(index, data, color='#F78F1E')
    plt.xlabel(x_label, fontsize=8)
    plt.ylabel(y_label, fontsize=8)
    plt.xticks(index, label, fontsize=8, rotation = 20)
    plt.title(title)
    plt.show()
```
